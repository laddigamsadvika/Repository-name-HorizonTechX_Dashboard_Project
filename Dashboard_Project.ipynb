```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
avg_math = df['math score'].mean()
avg_reading = df['reading score'].mean()
avg_writing = df['writing score'].mean()

print("Average Math Score:", round(avg_math, 2))
print("Average Reading Score:", round(avg_reading, 2))
print("Average Writing Score:", round(avg_writing, 2))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

subject_avg = df[['math score', 'reading score', 'writing score']].mean()
axes[0,0].bar(subject_avg.index, subject_avg.values)
axes[0,0].set_title("Subject-wise Average Scores")

gender_avg = df.groupby('gender')[['math score','reading score','writing score']].mean()
gender_avg.plot(kind='bar', ax=axes[0,1])
axes[0,1].set_title("Gender Performance Comparison")

prep_avg = df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
prep_avg.plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title("Impact of Test Preparation")

axes[1,1].scatter(df['reading score'], df['writing score'])
axes[1,1].set_title("Reading vs Writing Correlation")

plt.tight_layout()
plt.show()
```
